### visualization of scatter plot

```python
def scatterplot_Length1_vs_Weight(df):
    """the scatterplot displays relationship between Length1 and Weight.
    """
    plt.figure(figsize=(10,6))
    plt.scatter(df['Length1'],df['Weight'],alpha=0.8,edgecolor='blue')
    plt.title("Relationship b/w height1 and weight")
    plt.xlabel('Length1(standardized)')
    plt.ylabel('Weight(standardized)')
    plt.grid(True,linestyle="--",alpha=0.12)
    plt.tight_layout()
    plt.show()
    
#relationship plot
scatterplot_Length1_vs_Weight(df)
```

```python
def correlationship_of_heatmap(df):
    """The heatmap shows correlationship among numeric features in dataset.
    """
    plt.figure(figsize=(8,6))
    correlation_matrix=df.corr()
    sns.heatmap(correlation_matrix,annot=True, fmt='.2f',cmap='coolwarm',square=True,linewidths=0.4)
    plt.title('heatmap correlation')
    plt.tight_layout()
    plt.show()
    
correlationship_of_heatmap(df)
```

```python

import warnings
warnings.filterwarnings("ignore", category=UserWarning, message="KMeans is known to have a memory leak")

def elbow(df, features):

    distortions = []
    for k in range(1, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(df[features])
        distortions.append(kmeans.inertia_)

    # Creating the elbow plot
    plt.figure(figsize=(6, 6))  # Adjust figure size
    plt.plot(range(1, 11), distortions, marker='o', linestyle=' ', color='p', markersize=10)
    plt.title("Elbow Method for Optimal Number of Clusters", fontsize=14)
    plt.xlabel("Number of Clusters (K)", fontsize=14)
    plt.ylabel("Distortion (Inertia)", fontsize=14)
    plt.xticks(range(1, 11))  # Ensure all x-axis values are displayed
    plt.grid(True)
    plt.show()

# Usage
elbow(df, numerical_features)
```

### k-means clustering

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[numeric_features])

#  clusters with enhanced visualization OF CLUSTER
plt.figure(figsize=(8, 6))  # Adjust figure size
plt.scatter(df['Length1'], df['Height'], c=df['Cluster'], cmap='viridis', alpha=0.7, s=40)  # Increase marker size
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='X', s=200, label="Centroids")

# INSERTING LABELS 
plt.title("K-Means Clustering (Length1 vs Height)", fontsize=14)
plt.xlabel("Length1 (Standardized)", fontsize=14)
plt.ylabel("Height (Standardized)", fontsize=14)
plt.grid(True)
plt.legend(fontsize=14)

# Show the plot
plt.show()
```
